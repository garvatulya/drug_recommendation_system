Train Test Split

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder   # Used to convert Categorical data to numerical format
```

```python
X=dataset.drop("prognosis",axis=1)  # delete the column prognosis column wise, 1 represent column
```

```python
y=dataset["prognosis"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=20) # 70% data in train and 30% data in testing
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
le=LabelEncoder()
le.fit(y)  # The fit() method scans the input array y and identifies all unique labels in it. It learns and stores these unique labels, mapping each to a unique integer.
Y=le.transform(y) # The transform() method converts the original categorical labels in y into their corresponding integer values, as determined during the fit() step.The transform() method converts the original categorical labels in y into their corresponding integer values, as determined during the fit() step.
```

```python
Y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=20) # 70% data in train and 30% data in testing
```

Training Top Models

```python
from sklearn.datasets import make_classification
from sklearn.svm import SVC  # Support Vector Classifier,machine learning model used for classification tasks, and it implements the Support Vector Machine (SVM) algorithm.
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier # An ensemble model in machine learning combines multiple individual models (often called "base learners" or "weak learners") to create a more powerful and robust model. The main idea behind ensemble methods is that by combining the predictions of multiple models, you can reduce the risk of overfitting, improve generalization, and increase predictive performance.
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB  # This is a variant of the Naive Bayes classifier that is particularly well-suited for classification with discrete features, such as word counts or term frequencies in text data.

from sklearn.metrics import accuracy_score,confusion_matrix
import numpy as np

# Create a dictionary to store models

models = {
    "SVC": SVC(kernel='linear'),
    "Random Forest": RandomForestClassifier(n_estimators=100,random_state=42),
    "GradientBoosting": GradientBoostingClassifier(n_estimators=100,random_state=42),
    "KNeighbors": KNeighborsClassifier(n_neighbors=5),
    "MultinomialNB": MultinomialNB()
}

for model_name, model in models.items():
    # train model
    model.fit(X_train,y_train)  # common method in machine learning used to train a model on a dataset.

    # test model
    predictions=model.predict(X_test)

    # Calculate accuracy
    accuracy=accuracy_score(y_test,predictions)

    # Calculate confusion matrix
    cm=confusion_matrix(y_test,predictions)

    print(f"{model_name} Accuracy: {accuracy}")
    print(f"{model_name} Confusion Matrix:")
    print(np.array2string(cm,separator=', '))
```

Single Prediction

```python
svc=SVC(kernel='linear')
svc.fit(X_train,y_train)
ypred=svc.predict(X_test)
accuracy_score(y_test,ypred)
```
